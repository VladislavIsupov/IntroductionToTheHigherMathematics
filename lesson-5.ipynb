```python
x = np.random.randint(0, 37, 1000000)
m = 0
n = 0
z = 0
for x_i in x:
    n += 1
    if x_i == 15 or x_i == 17:
        m += 1
    elif x_i == 0:
        z += 1
print (f'Экспериментальная вероятность выпадения числа 15 или 17 - {m/n:0.4}')
print (f'Вероятность выпадения или 15, или 17 равна сумме вероятностей каждого события : {1/37 + 1/37:0.4}')
print (f'Экспериментальная вероятность выпадения Zero : {z/n:0.4}')
print (f'Расчетная вероятность : {1/37:0.4}')
        
        
```

```python
k, n, m = 0, 5, 100000

x = np.random.randint(0, 37, (m, n))

for i in range(0, m):
    k_i = 0
    for j in range(0, n):
        if (x[i, j] == 15 or x[i, j] == 17):
            k_i += 1
    if k_i == 2:
        k += 1
print(f'Количество успехов: {k}, количество выборок: {m}, \n \
      вероятность выпадения 15 или 17 в 5-ти испытаниях два раза: {k/m}')
```

```python
k, n, m = 0, 13, 100000

x = np.random.randint(0, 37, (m, n))

for i in range(0, m):
    k_i = 0
    for j in range(0, n):
        if (x[i, j] == 15 or x[i, j] == 17):
            k_i += 1
    if k_i == 3:
        k += 1
print(f'Количество успехов: {k}, количество выборок: {m}, \n \
      вероятность выпадения 15 или 17 в 13-ти испытаниях три раза: {k/m}')
```
