```python
x = np.linspace(-10, 10, num=101, endpoint=True)
```

```python
y = np.cos(x-1)-1
y2 = 0.5*np.cos(x+2)+1
y3 = 1.5*np.cos(x-1)+0.5
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "Gray",
    "family": "serif",
}

label_font = {
    "fontsize": 10,
    "family": "serif",
    "color": "#505050",
}

legend_font = {
    "size": 7,
    "family": "serif",
}
```

```python
plt.plot(x, y, color = 'green', label = 'cos(x-1)-1')
plt.plot(x, y2, color = 'blue', label = '0.5*cos(x+2)+1')
plt.plot(x, y3, color = 'red', label = '1.5*cos(x-1)+0.5')

plt.axis([-9.5, 9.5, -2.7, 2.7])

plt.title('График y(k,x)', fontdict=title_font)
plt.xlabel('Значения x', fontdict=label_font)
plt.ylabel('Значения y', fontdict=label_font)

plt.grid(axis="both", color="lightgrey", linestyle = 'dashed')

legend = plt.legend(loc = 'upper left', frameon=False, prop=legend_font)
plt.setp(legend.get_texts(), color="Black")
```

```python
from scipy.optimize import fsolve
```

```python
def equations(p):
    x, y = p
    return(y-x**2+1, np.exp(x)+x*(1-y)-1)
```

```python
x1, y1 = fsolve(equations, (-2, 3))
x2, y2 = fsolve(equations, (2.5, 5))
x3, y3 = fsolve(equations, (4, 15))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```
