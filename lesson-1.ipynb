```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-10, 10, num=101, endpoint=True)
x
```

```python
k = 1
k2 = .5
```

```python
y = np.cos(k*x)
y2 = np.cos(k2*x)
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "Gray",
    "family": "serif",
}

label_font = {
    "fontsize": 10,
    "family": "serif",
    "color": "#505050",
}

legend_font = {
    "size": 7,
    "family": "serif",
}
```

```python
plt.plot(x, y, color = 'green', label = 'cos(1*x)')
plt.plot(x, y2, color = 'blue', label = 'cos(0.5*x)')

plt.axis([-9.5, 9.5, -1.3, 1.3])

plt.title('График y(k,x)', fontdict=title_font)
plt.xlabel('Значения x', fontdict=label_font)
plt.ylabel('Значения y', fontdict=label_font)

plt.grid(axis="both", color="lightgrey", linestyle = 'dashed')

legend = plt.legend(loc = 'upper left', frameon=False, prop=legend_font)
plt.setp(legend.get_texts(), color="Gray")

plt.show
```
