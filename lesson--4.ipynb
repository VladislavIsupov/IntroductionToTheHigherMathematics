```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
rcParams["figure.figsize"] = 8, 5
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "Gray",
    "family": "serif",
}

label_font = {
    "fontsize": 14,
    "family": "serif",
    "color": "#505050",
}

legend_font = {
    "size": 10,
    "family": "serif",
}
```

```python
x = np.linspace(100, 500, 100)
n = 2

# Построим графики решения
plt.plot(x, (1 * np.pi/x), label="n=1", color="green")
plt.plot(x, (2 * np.pi/x), label="n=2", color="blue")
plt.plot(x, (3 * np.pi/x), label="n=3", color="red")

# Установим границы графика
plt.axis([100, 500, .01, .02])

# Отобразим сетку
plt.grid(axis="both", color="lightgrey", linestyle = 'dotted')

# Заголовок графика и названия осей
plt.title('Графики a(n,x)', fontdict=title_font)
plt.xlabel('Значения x', fontdict=label_font)
plt.ylabel('Значения a', fontdict=label_font)

# Сетка
plt.grid(axis="both", color="lightgrey", linestyle = 'dashed')

# Легенда
legend = plt.legend(loc = 'lower left', frameon=False, prop=legend_font)
plt.setp(legend.get_texts(), color="Black")
```
